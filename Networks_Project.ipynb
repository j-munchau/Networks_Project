#### Background
A network in its simplest form is a graph representation of a real-world system in which nodes are entities of 
interest (e.g., individuals) and edges are pairwise interactions between entities (e.g., connections on social media). 
Networks are useful representations in a wide range of applications, such as sociology, biology, economics, transportation, and so on. 

A broad range of network features can be characterised via properties of matrices commonly used to describe networks. *Communities* or *clusters* are a very commonly studied type of network feature. These are loosely defined as sets of nodes that are strongly connected in a network (e.g., echo chambers in a Twitter social network, brain regions contributing to similar functions in a neuronal network). In this project, we will explore how two matrices known as the *adjacency matrix* and the *Laplacian* can be used to decompose real-world networks into communities.

#### A2a
What does the function below return for each node $i$ of the network? 

Add a markdown cell below the code cell below in which to answer this question.

```python
def deg_node(A,i):
    k = 0
    for j in range(A.shape[0]):
        k = k + A[i][j]
    return k
```

For each node i of the network, this function returns the number of edges connected to it. This is because for each node j that node i is connected to, the corresponding entry in the adjacency matrix A, $a_{ij}$, will be 1 and if there's no edge between nodes i and j, the entry in the adjacency matrix will be 0. So, since the function sums all entries in row i of the adjacency matrix, it will sum to the total number of edges connected to node i.

### Task A3: Laplacian [approximately 20%]
The *Laplacian* $L$ of a symmetric adjacency matrix $A$ is defined as $L = D - A$, where $D$ is a diagonal matrix with $(i,i)$th entry the degree $k_i$ of node $i$. The Laplacian plays many important roles in understanding the structure of networks, and one such role is in understanding the community structure of networks.

```python
def get_diagonal(A): #function to get the diagonal matrix from an adjacency matrix
    D = np.zeros((A.shape[0],A.shape[0])) #generate a 34x34 matrix of zeros
    for i in range(A.shape[0]):
        D[i][i] = deg_node(A,i) #makes the (i,i)th entry the degree of the ith node
    return D #returns the diagonal matrix

def get_laplacian(A):
    diagonal_matrix = get_diagonal(A)
    L = np.subtract(diagonal_matrix,A) #subtracts the adjacency matrix from the diagonal matrix
    return(L)

def sum_column(A,i): #gets the sum of each column
    tot = 0
    for j in range(A.shape[0]):
        tot = tot + get_laplacian(A)[j][i]
    return(tot)

def sum_row(A,i): #gets the sum of each row
    tot = 0
    for j in range(A.shape[0]):
        tot = tot + get_laplacian(A)[i][j]
    return(tot)

def sums_of_rows(A): #prints the sum of all rows 
    for i in range(A.shape[0]):
        print("The sum of each entry in row", i, "is", sum_row(A,i))
        
def sums_of_columns(A): #prints the sum of all columns
    for i in range(A.shape[0]):
        print("The sum of each entry in column", i, "is", sum_column(A,i))
```

In the diagonal matrix, the only no-zero entries are along the leading diagonal and are the degree of each node, i.e the entry (i,i) is the sum of the entries in row/column i of the adjacency matrix. By subtracting the adjacency matrix from the diagonal matrix, the sum of each row and column will be equal to zero since you are subtracting the degree of each node from the degree of each node

#### A4a
Suppose we have a partition of a network into two sets of nodes. The number of edges $R$ falling between these sets is called the cut size and is given by 
\begin{equation*}
R = \frac{1}{2}\sum_{\substack{i,j \text{ in} \\\text{different}\\ \text{sets}}} A_{ij}\,.

Show that $R$ can be rewritten as $R = \frac{1}{4}s^TLs$ where

Using the signs of the Fiedler eigenvector, one can write a heuristic algorithm that splits the nodes in the Karate Club into two clusters. One can then encode the clustering using a partition array.

#### Answer to A4a
Suppose we have $N$ nodes, then the adjacency matrix is $N\times N$. We will call the two sets of nodes $C_1$ (Set 1) and $C_2$ (Set 2) and $V$ the set of all nodes, (We have $C_1\cup C_2=V$ and $C_1\cap C_2=\emptyset$). Therefore our $s_i$ becomes:
$$
s_i=\begin{cases}+1\qquad i\in C_1\\ -1\qquad i\in C_2\end{cases}
$$
So if we take any two nodes $i$ and $j$ we have
$$
\frac{1}{2}(1-s_is_j)=\begin{cases}1\quad i,j\text{ in different sets}\\0\quad\text{otherwise} \end{cases}
$$
($i,j$ being in different set means either $i\in C_1$ and $j\in C_2$ or $i\in C_2$ and $j\in C_1$) Hence we can rewrite $R$ as
\begin{align*}
R&=\frac{1}{4}\sum_{i,j\in V}(1-s_is_j)A_{ij}\\
&= \frac{1}{4}\sum_{i,j\in V} A_{ij}-s_is_jA_{ij}\\
&=\frac{1}{4}\left(\sum_{i,j\in V}A_{ij}-\sum_{i,j\in V}s_is_jA_{ij}\right)
\end{align*}
We have that
\begin{align*}
\sum_{i,j\in V}A_{ij}=\sum_{i\in V}\sum_{j\in V}A_{ij}=\sum_{i\in V}k_i=\sum_{i,j\in V}k_i\delta_{ij}=\sum_{i,j\in V}s_is_jk_i\delta_{ij}
\end{align*}
(Since when $i=j$, we have that $\delta_{ij}=1$ and $s_is_j=1$ and otherwise we have that $\delta_{ij}=0$). Hence we have
\begin{align*}
R&=\frac{1}{4}\left(\sum_{i,j\in V}A_{ij}-\sum_{i,j\in V}s_is_jA_{ij}\right)\\
&=\frac{1}{4}\left(\sum_{i,j\in V}s_is_jk_i\delta_{ij}-\sum_{i,j\in V}s_is_jA_{ij}\right)\\
&=\frac{1}{4}\sum_{i,j\in V}s_is_jk_i\delta_{ij}-s_is_jA_{ij}\\
&=\frac{1}{4}\sum_{i,j\in V}s_is_j\left(k_i\delta_{ij}-A_{ij}\right)
\end{align*}
We can see that $L_{ij}=k_i\delta_{ij}-A_{ij}$ is actually the element of the Laplacian matrix of $A$ in row $i$ and column $j$ ($k_i\delta_{ij}$ represent the diagonal matrix where the $(i,i)\text{th}$ entry is the degree $k_i$ of node $i$). So
$$
R=\frac{1}{4}\sum_{i,j\in V}s_is_jL_{ij}=\frac{1}{4}\sum_{i\in V}s_i\sum_{j\in V}L_{ij}s_j
$$
If we let $s=(s_i)$ (column vector) then we can see that $\sum_j L_{ij}s_j$ for some $i$ is just the $i\text{th}$ row of $L$ dot producted with $s$, so $ s_i\sum_j L_{ij}s_j$ for some $i$ is just the $i\text{th}$ row of $L$ dot producted with $s$ and then multiplied by the $i\text{th}$ row of $s$. Hence we have that
$$
R=\frac{1}{4}\sum_{i\in V}s_i\sum_{j\in V}L_{ij}s_j=\frac{1}{4}s^TLs
$$
(What we are doing is dot producting each row of $L$ $\left(N\times N\text{ matrix}\right)$ with $s$ $\left(N\times1\text{ column vector}\right)$ which is equivalent to $Ls$ which is an $N\times 1$ matrix, and then since each row of $Ls$ is multiplied by the same row in $s$ and then the rows are summed, we are doing a dot product of $s$ and $Ls$ which is equivalent to the matrix multiplication $s^TLs$)

```python
def spectral_bipartitioning(A, k_seq):
    laplacian_matrix = np.diagflat(k_seq) - A
    eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)
    index_1 = np.argsort(eigenvalues)[1]
    partition = [val >= 0 for val in eigenvectors[:, index_1]]
    partition_array = 1*np.reshape(partition, A.shape[0])
    return partition_array
```

#### Answer to A4b
This function takes in two parameters, the Adjacency matrix as $A$ and an array k_seq which is an array containing the degree of each node in order i.e. the degree of the $i\text{th}$ node is in k[i-1].

The first line of the function creates the Laplacian Matrix, where the numpy function diagflat creates a diagonal matrix given an array which in this case is the array containing the degrees of the nodes and therefore the line represents the equation $L=D-A$. 
The second line of the function uses the numpy function linalg.eigh which return a namedtuple containing an array of the eigenvalues and another array (2D) of the eigenvectors of the adjacency matrix $A$ and these arrays are assigned to variables namd eigenvalues and eigenvectors respectively. 

The third line of the function assigns np.argsort(eigenvalues)[1] to the variable index_1. The function np.argsort() returns of an array that contains the indicies that sort the array (in an ascending order) i.e. replacing each index by it's corresponding value gives a sorted array. Hence np.argsort(eigenvalues)[1] returns the index of the second least eigenvalue. Since we know from A3c that 0 is always an eigenvalue of the Laplacian, and that the eigenvalues of the Laplacian are always non-negative then the second least eigenvalue is the least non-zero eigenvalue. 

In the fourth line of the function, eigenvectors[:, index_1] outputs the index_1th row i.e. outputs the eigenvector corresponding to the least non-zero eigenvalue. Therefore [val >= 0 for val in eigenvectors[:, index_1]] loops through each element in the index_1th row and creates a new boolean array (linked list) of the same length, where each element is True if the corresponding element in the index_1th row is non-negative and false if it is negative. This array is then assigned to the variable partition.

The fifth line of the function, converts the partition linked list into a numpy array (the dimension of partition is A.shape[0]) and then by multiplying by 1, each value in the Boolean array that is True is changed to 1, and each value that is False is changed to 0.

The output array only contains either a 1, or a 0. The indexes of the output array correspond to each node and are put into clusters depending on the value at that index, i.e. all indexes that have the value 1 in the output array are the nodes of one cluster, and the other cluster is the nodes corresponding to the indexes that have value 0 in the output array.
